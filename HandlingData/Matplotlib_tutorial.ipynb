## Importing Library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Line Plot

```python
x = [2, 3 , 6, 7, 8]
y = [3, 7 , 4, 12, 5]
```

```python
plt.plot(x,y)
plt.title("Line graph")
plt.ylabel('y-axis')
plt.xlabel('x-axis')
plt.show()
```

```python
plt.plot(x,y)
plt.plot(x,y, '*')
plt.title("Line graph")
plt.ylabel('y-axis')
plt.xlabel('x-axis')
plt.show()
```

```python
x = np.arange(0,3*np.pi,0.1)
y = np.sin(x)
```

## Scatter Plot

```python
x = [2, 3 , 6, 7, 8]
y = [3, 7 , 4, 12, 5]
```

```python
plt.scatter(x,y)

plt.title("Scatter plot")
```

```python
x2 = [1, 2, 3, 4, 5]
y2 = [2, 4,6,8,10]
```

## Comparison

```python
plt.plot(x,y, color='g', linewidth = 6)
plt.plot(x2,y2, color='r', linewidth = 4)
plt.title("Comparison graph")

plt.annotate(xy=[2,3], s="First line")
plt.annotate(xy=[5,10], s="Second line")

plt.legend(['First line','Second line'], loc=4)
plt.show
```

## Vertical Bar Graph

```python
x = [2, 3 , 6, 7, 8]
y = [3, 7 , 4, 12, 5]
x2 = [1, 2, 3, 4, 5]
y2 = [2, 4,6,8,10]
```

```python
plt.bar(x,y, color='g', linewidth = 6)
plt.bar(x2,y2, color='r', linewidth = 4)
plt.title('Bar graphs')
```

## Horizontal Bar Graph

```python
plt.barh(x,y, color='g', linewidth = 6)
plt.barh(x2,y2, color='r', linewidth = 4)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Bar graphs')
```

### Bubble Graph

## Pie Chart

```python
data = [3,7,4,12,5]
lb= ['a','b','c','d','e']
plt.pie(data, labels=lb)
plt.show
```

```python
ex = [0,0.1,0,0,0]
plt.pie(data,labels=lb, explode=ex, shadow=True)
plt.show
```
