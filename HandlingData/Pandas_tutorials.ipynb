## Missing Values


```python
data = {
    'roll_no' : [3,2,7,11],
    'ppr_id' : [43,52,50,20],
    'marks' : [np.nan,23,17,27]
}
```

```python
df1 = pd.DataFrame(data)
```

```python
df1
```

### isnull()

```python
df1.isnull()
```

```python
df1.isnull().sum()
```

### fillna()

```python
df2 = df1.fillna(1) # fill null values with 1
```

```python
df2
```

### Dropping null values


```python
a = df1.dropna() # drops the entire row
```

```python
df1
```

```python
a
```

```python
a = df1.dropna(axis = 0) # dropping the row
a
```

```python
a = df1.dropna(axis=1) # dropping the column
a
```

### Creating a data with non NULL values

```python
a = pd.notnull(df1["marks"])
a
```

```python
df1[a]  
```
