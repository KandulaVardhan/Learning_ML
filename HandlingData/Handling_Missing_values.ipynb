### Using Imputer

Strategy to handle missing values can be: mean, median, most_frequent

```python
# automatically uses mean / median / most frequent values to replace the missing data without changing the distribution
from sklearn.preprocessing import Imputer
imputer_mean = Imputer(missing_values = "NaN", strategy="mean", axis = 0)
imputer_median = Imputer(missing_values = "NaN", strategy="median", axis = 0)

imputer_data_mean = imputer_mean.fit_transform((data[cols]))
imputer_data_median = imputer_median.fit_transform(data[cols])

imputer_data_mean = pd.DataFrame(data=imputer_data_mean, columns=cols)
imputer_data_median = pd.DataFrame(data=imputer_data_median, columns =cols)
```
